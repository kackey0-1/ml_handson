# 標本の統計量と性質

```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%precision 3
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
population = stats.norm(loc=4, scale=0.8)
# loc=期待値, scale=分散
population
```

## 標本平均の平均値は母平均に近似する

- シミュレーション
  - 対象: 標本平均
  - 調べたいこと: Nを大きくする(標本数を増やす)と標本平均の平均は母平均に近似するか

```python
np.random.seed(1)
N = 10000
sample_mean_array = np.zeros(N)
for i in range(0, N):
    sample = population.rvs(size=10)
    sample_mean_array[i] = sp.mean(sample)

print(f'標本平均配列:      {sample_mean_array[0:10]}')
print(f'標本平均の平均:    {sp.mean(sample_mean_array)}')
print(f'標本平均の分散:    {sp.var(sample_mean_array, ddof=1)}')
print(f'標本平均の標準偏差: {sp.std(sample_mean_array, ddof=1)}')
sns.histplot(sample_mean_array,)
```

## サンプルサイズが十分に大きい場合、標準平均は母平均に近似する

- シミュレーション
  - 対象: 標本平均
  - 変化させるもの: サンプルサイズ
  - 調べたいこと: サンプルサイズが大きくなると、標準平均は母平均に近づくか

```python
size_array = np.arange(start=10, stop=100100, step=100)
sample_mean_array_size = np.zeros(len(size_array))

np.random.seed(1)
for i in range(0, len(size_array)):
    sample = population.rvs(size=size_array[i])
    sample_mean_array_size[i] = sp.mean(sample)

print(f'sample_size(x): {size_array[0::]}')
print(f'sample_mean(y): {sample_mean_array_size[0::]}')

plt.plot(size_array, sample_mean_array_size)
plt.xlabel('sample size')
plt.xlabel('sample mean')
```

```python
# サンプルサイズを変えた時の標本平均の分布
def calc_sample_mean(sample_size, n_trial):
    sample_mean_arr = np.zeros(n_trial)
    for i in(range(0, n_trial)):
        s = population.rvs(size=sample_size)
        sample_mean_arr[i] = sp.mean(s)
    return sample_mean_arr

np.random.seed(1)
size_10 = calc_sample_mean(sample_size = 10, n_trial=10000)
print(size_10)
size_10_df = pd.DataFrame({
    "sample_mean": size_10,
    "size": sp.tile("size 10", 10000)
})
size_20 = calc_sample_mean(sample_size = 20, n_trial=10000)
size_20_df = pd.DataFrame({
    "sample_mean": size_20,
    "size": sp.tile("size 20", 10000)
})
size_30 = calc_sample_mean(sample_size = 30, n_trial=10000)
size_30_df = pd.DataFrame({
    "sample_mean": size_30,
    "size": sp.tile("size 30", 10000)
})

sim_result = pd.concat([size_10_df, size_20_df, size_30_df])
# print(sim_result)
sns.violinplot(data=sim_result, x="size", y="sample_mean")
# サンプルサイズが大きくなると、標本平均のばらつきが小さくなることがわかる
```

## 標本平均の標準偏差は、母標準偏差よりも小さい

- シミュレーション
  - 対象: 標本平均の標準偏差
  - 変化させるもの: サンプルサイズ
  - サンプルが大きくなると、標準平均の標準偏差は小さくなる -> サンプルサイズを大きくすると標準平均はより信頼性が増す

```python
size_array = np.arange(start=2, stop=102, step=2)
print(size_array[0::])
sample_mean_std_array = np.zeros(len(size_array))

for i in range(0, len(size_array)):
    sample_mean = calc_sample_mean(
        sample_size=size_array[i],
        n_trial=100
    )
    sample_mean_std_array[i] = sp.std(sample_mean, ddof=1)

plt.plot(size_array, sample_mean_std_array)
plt.xlabel('sample size')
plt.ylabel('mean std value')
```

### 標準誤差(SE=Standard Error)

理論上の標本平均の標準偏差の大きさを指す(=標準誤差)

$$
標準誤差 = \frac{\sigma}{\sqrt{N}}
$$

```python
standard_error = 0.8 / np.sqrt(size_array)
print(standard_error[0::])

plt.plot(size_array, sample_mean_std_array) # 標本平均の標準偏差
plt.plot(size_array, standard_error, linestyle='dotted') # 標準誤差
plt.xlabel('sample size')
plt.ylabel('mean std value')

# 標準誤差と標本平均の標準偏差の値がほぼ一致していることがわかる
```

## 標本分散の平均値は、母分散からズレている/不偏分散を使うとバイアスがなくなる

- シミュレーション
  - 対象: 標本分散
  - 調査するもの: 標本分散の平均値は、母分散からずれるか


```python
N = 10000
sample_var_array = np.zeros(N)
unbias_var_array = np.zeros(N)
for i in range(0, N):
    sample = population.rvs(size=10)
    # ddof=0 -> 標本分散
    # ddof=1 -> 不偏分散
    sample_var_array[i] = sp.var(sample, ddof=0)
    unbias_var_array[i] = sp.var(sample, ddof=1)



print(f'標本分散: {sp.mean(sample_var_array)}')
print(f'不偏本分散: {sp.mean(unbias_var_array)}')
```

## サンプルサイズが十分に大きい場合、不偏分散は母平均に近似する

- シミュレーション
  - 対象: 不偏分散
  - 変化させるもの: サンプルサイズ
  - 調べたいこと: サンプルサイズが大きくなると不偏分散は母分散に近似する

```python
size_array = np.arange(start=10, stop=100100, step=100)
sample_unbias_var_array = np.zeros(len(size_array))

np.random.seed(1)
for i in range(0, len(size_array)):
    sample = population.rvs(size=size_array[i])
    sample_unbias_var_array[i] = sp.var(sample, ddof=1)

print(f'sample_size(x)    : {size_array[0::]}')
print(f'unbias variance(y): {sample_unbias_var_array[0::]}')

plt.plot(size_array, sample_unbias_var_array)
plt.xlabel('sample size')
plt.xlabel('unbias variance')

# 推定量の期待値が真の母数(母集団パラメタ)となる特性を不偏性と呼ぶ
# サンプルサイズが大きくなると、推定量が真の母数に近似することを一致性と呼ぶ
```

## 母数に対する良い推定量

- 分析の目的は、基本的に母集団分布を推定することにある
- 母集団分布がわかれば、未知のデータに対して、予測・推測を行うことが可能
- 例
  - 母集団分布に正規分布を仮定すると、正規分布の母数を推定うすることで、母集団分布を推定できる
    - 正規分布の母数は平均と分散の2つ
    - 正規分布の母数はわからないため、不偏正を持つ、`標本平均`と`不偏分散`を用いる


* 不偏性: 推定量の期待値が真の母数(母集団パラメタ)となる特性を不偏性と呼ぶ
* 一致性: サンプルサイズが大きくなると、推定量が真の母数に近似することを一致性と呼ぶ


## 大数の法則

大数の法則は、*標本の大きさが大きくなるにつれて、標本平均が母平均に近づく近づき方を表現した法則*


## 中心極限定理

母集団分布が何であっても、サンプルサイズが十分に大きい時には、確率変数の和は正規分布に近いものになる

- シミュレーション
  - 対象: 確率変数の和
  - 変化するもの: サンプルサイズ
  - 調べるもの: サンプルサイズを大きくすると、確率変数の和は正規分布に近いものになるか

```python
n_size = 10000
n_trial = 50000

coin = np.array([0, 1])
count_coin = np.zeros(n_trial)
np.random.seed(1)

for i in range(0, n_trial):
    count_coin[i] = sp.sum(
        np.random.choice(coin, size=n_size, replace=True)
    )

print(np.random.choice(coin, size=n_size, replace=True))
print(count_coin[1])
sns.distplot(count_coin)
```
